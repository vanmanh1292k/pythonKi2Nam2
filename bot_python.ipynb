```python
class main:
    def __init__(self):
        self.condition = 'A'

    def look(self):
        if self.condition == 'A':
            self.condition = 'A'
            return 2
        elif self.condition == 'C':
            self.condition = 'F'
            return 5
        elif self.condition == 'D':
            self.condition = 'E'
            return 6
        elif self.condition == 'F':
            self.condition = 'F'
            return 8
        else:
            raise KeyError

    def sit(self):
        if self.condition == 'A':
            self.condition = 'B'
            return 0
        elif self.condition == 'E':
            self.condition = 'F'
            return 7
        else:
            raise KeyError
    def fork(self):
        if self.condition == 'A':
            self.condition = 'E'
            return 1
        elif self.condition == 'B':
            self.condition = 'C'
            return 3
        elif self.condition == 'C':
            self.condition = 'D'
            return 4
        else:
            raise KeyError

o = main()
print(o.look())
print(o.look())
print(o.sit())
print(o.fork())
print(o.fork())
print(o.fork())
print(o.look())
print(o.look())
print(o.sit())
print(o.look())

```

```python
import math
import collections
import datetime as DT
table = [['.. ',' ','19/11/03#0.954'],
             ['.. ',' ','14/03/03#0.936'],
             ['.. ',' ','18/05/02#0.404']]
table1 = [['.. ',' ','09/03/99#0.064'],
             ['.. ',' ','14/03/03#0.027'],
             ['.. ',' ','02/05/04#0.880']]

table2 = [[None, None, None], [None, None, None], ['.. ', None, '09/03/99#0.064'], ['.. ', None, '14/03/03#0.027'], ['.. ', None, '02/05/04#0.880']]


def main(table):

    while [None, None, None] in table:
        table.remove([None, None, None])

    for i in range(0, len(table)):
        table[i].pop(1)

    for i in range(len(table)):
        x = table[i][0].split('.')
        table[i][0] = x[2].strip() + ' ' + x[0] + '.'
        for j in range(len(table[i])):
            if '#' in table[i][j]:
                temp = table[i][j]
                table[i][j] = str(round(float(temp[9:14]), 1))
                table[i].append(temp[0:8])

    for i in range(len(table)):
        date = DT.datetime.strptime(table[i][2], '%d/%m/%y').date()
        table[i][2] = date.strftime('%y/%m/%d')
    table = [list(x) for x in zip(*table)]
    return table

print(main(table2))
```
